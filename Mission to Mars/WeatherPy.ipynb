```python

import pandas as pd
```

```python
# Importing city data generated in WeatherPy
city_data_path = "weather_data.csv"
city_data_read = pd.read_csv(city_data_path)
city_data_cleand_df = city_data_read.drop(columns=["Unnamed: 0"])
city_data_cleand_df
```

```python
# Reordering columns
city_data_cleand_df = city_data_cleand_df[["City", 
                                           "Country", 
                                           "Latitude", 
                                           "Longitude", 
                                           "Temperature (F)", 
                                           "Humidity (%)", 
                                           "Cloud Cover (%)", 
                                           "Wind Speed (MPH)"]]

city_data_cleand_df
```

```python
# Gives me data table in HTML
print(city_data_cleand_df.to_html())
```
